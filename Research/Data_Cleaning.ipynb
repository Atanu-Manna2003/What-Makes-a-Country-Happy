```python
import pandas as pd
import numpy as np
```

```python
path2015= r"F:\What-Makes-a-Country-Happy\Data\2015.csv"
 # Load dataset
df2015 = pd.read_csv(path2015)
df2015.columns
```

```python
# Rename the columns
df2015["Year"] = 2015
df2015 = df2015.rename(columns={
    "Family": "Social Support",
    "Economy (GDP per Capita)": "Economy",
    "Health (Life Expectancy)": "Life Expectancy",
    "Trust (Government Corruption)": "Trust On Govt"
})
df2015=df2015.drop(columns=["Standard Error"])
df2015.head()
```

```python
path2016= r"F:\What-Makes-a-Country-Happy\Data\2016.csv"
# Load dataset
df2016 = pd.read_csv(path2016)
df2016.columns

```

```python
# Rename the columns
df2016["Year"] = 2016
df2016 = df2016.rename(columns={
    "Family": "Social Support",
    "Economy (GDP per Capita)": "Economy",
    "Health (Life Expectancy)": "Life Expectancy",
    "Trust (Government Corruption)": "Trust On Govt"
})
df2016=df2016.drop(columns=["Lower Confidence Interval","Upper Confidence Interval"])
df2016.head()
```

```python

path2017= r"F:\What-Makes-a-Country-Happy\Data\2017.csv"
# Load dataset
df2017 = pd.read_csv(path2017)
df2017.columns
```

```python

# Add a new column with null (NaN) values
df2017["Year"] = 2017
df2017["Region"] = np.nan
# Rename the columns
df2017 = df2017.rename(columns={
    "Happiness.Rank":"Happiness Rank",
    "Happiness.Score":"Happiness Score",
    "Economy..GDP.per.Capita.":"Economy",
    "Family": "Social Support",
    "Health..Life.Expectancy.": "Life Expectancy",
    "Trust..Government.Corruption.": "Trust On Govt",
    "Dystopia.Residual":"Dystopia Residual"
})
df2017=df2017.drop(columns=["Whisker.high","Whisker.low"])
df2017.head()
```

```python
path2018= r"F:\What-Makes-a-Country-Happy\Data\2018.csv"
# Load dataset
df2018 = pd.read_csv(path2018)
df2018.columns
```

```python
# Drop the column
df2018["Year"] = 2018
# Add a new column with null (NaN) values
df2018["Region"] = np.nan
df2018["Dystopia Residual"] = np.nan
# Rename the columns
df2018 = df2018.rename(columns={
    "Country or region":"Country",
    "Overall rank":"Happiness Rank",
    "Score":"Happiness Score",
    "GDP per capita":"Economy",
    "Social support": "Social Support",
    "Healthy life expectancy": "Life Expectancy",
    "Freedom to make life choices":"Freedom",
    "Perceptions of corruption": "Trust On Govt"
   })

df2018.head()
```

```python

path2019= r"F:\What-Makes-a-Country-Happy\Data\2019.csv"
# Load dataset
df2019 = pd.read_csv(path2019)
df2019.columns
```

```python
# Drop the column
df2019["Year"] = 2019
# Add a new column with null (NaN) values
df2019["Region"] = np.nan
df2019["Dystopia Residual"] = np.nan
# Rename the columns
df2019 = df2019.rename(columns={
    "Country or region":"Country",
    "Overall rank":"Happiness Rank",
    "Score":"Happiness Score",
    "GDP per capita":"Economy",
    "Social support": "Social Support",
    "Healthy life expectancy": "Life Expectancy",
    "Freedom to make life choices":"Freedom",
    "Perceptions of corruption": "Trust On Govt"
   
})

df2019.head()
```

```python
# Merge (concatenate) all into a single dataframe
final_df = pd.concat([df2015, df2016, df2017, df2018, df2019], ignore_index=True)

final_df
```

```python
# Group by Country and count rows in each group
country_counts = final_df.groupby("Country").size().reset_index(name="Count")
country_counts
```

```python
# Step 1: Count countries
country_counts = final_df["Country"].value_counts()

# Step 2: Keep only countries with count = 5
valid_countries = country_counts[country_counts == 5].index

# Step 3: Filter the dataframe
filtered_df = final_df[final_df["Country"].isin(valid_countries)]
filtered_df
```

```python
filtered_df.isnull().sum()
```

```python
filtered_df["Trust On Govt"] = filtered_df["Trust On Govt"].fillna(filtered_df["Trust On Govt"].median())
```

```python
filtered_df.isnull().sum()
```

```python


# Fill missing Region values by grouping on Country
region_map = (
    filtered_df.groupby("Country")["Region"]
    .apply(lambda x: x.dropna().iloc[0] if x.notna().any() else None)
)

filtered_df["Region"] = filtered_df.apply(
    lambda row: region_map[row["Country"]] if pd.isna(row["Region"]) else row["Region"],
    axis=1
)
```

```python
filtered_df.isnull().sum()
```

```python
# Fill missing Dystopia Residual
columns_to_sum = ["Economy", "Social Support", "Life Expectancy", "Freedom", "Generosity","Trust On Govt"]

filtered_df["Dystopia Residual"] = filtered_df.apply(
    lambda row: row["Happiness Score"] - row[columns_to_sum].sum() 
                if pd.isna(row["Dystopia Residual"]) else row["Dystopia Residual"],
    axis=1
)
filtered_df.isnull().sum()
```

```python
filtered_df.to_csv("Cleand_Dataset_World_Happiness.csv",index=False)
```
